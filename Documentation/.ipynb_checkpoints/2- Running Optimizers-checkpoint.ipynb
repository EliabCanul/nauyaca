## 2.- Runing the optimizers

Optimizers are helpful to quickly explore the parameter space and search approximate solutions to the TTVs through minimizing a chi square statistic.

Optimizers module join a chain of methods namely, Differential Evolution, Powell and Nelder-Mead minimization methods to track solutions that best reproduce the TTVs.

```python
import nauyaca_local as nau
from nauyaca_local import Optimizers
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

In the Introduction we create a Planetary system object that was saved as 'SystemX.pkl'. Let's load it since it contains all the information about the planetary system:

```python
PS = nau.utils.load_pickle('SystemX.pkl')
print(PS)
```

```python
# Reference time of the solutions
PS.T0JD = 0.0

print(PS.T0JD)
```

## 2.1 Running the optimizers

Now is time to run the optimizers. To run it, jus call .run_optimizers and pass the Planetary System object created previously, set the number of solutions (nsols) and specify the number of cores to run in parallel (cores).

```python
# Run Optimizers. Results of this method are saved in a file  with planetary system name + '.opt'

RESULTS_opt = Optimizers.run_optimizers(PS, nsols=160, cores=8)
```

```python
RESULTS_opt = np.genfromtxt(f'{PS.system_name}.opt').tolist()
```

Results have been saved in a .opt file in the current directory, you can take a look! All the solutions from the optimizer consist in a list, being the first element the $\chi^2$ and the rest are the planetary parameters of the first planet concatenated with planetary parameters of the next planet(s). A flag of chi square of 1e+50 could rise if the minimizers could not find a solution.

```python
# The first 3 results (unordered by chi2)
print(RESULTS_opt[:3])
```

## 2.2 Visualizing the results

Let's plot some results, but first let's define a convenient dictionary of the parameter labels using the .params_names_opt attribute of the Planetary system

```python
parameters = {}
for i, names in enumerate(PS.params_names_opt.split()):
    parameters[names] = i

print(parameters)
```

```python
def plot_opt(param, color, bins):
    plt.figure()
    values = [i[parameters[param]+1] for i in RESULTS_opt]  # +1 to avoid chi**2
    sns.distplot(values, kde=False, color=color, bins=bins)
    plt.xlabel(f'{param}')
    

# Mass of planet b (Mass1)
param = 'mass1'
plot_opt(param, 'blue', 10)

# Eccentricity of planet c (ecc2)
param = 'mass2'
plot_opt(param, 'orange', 10)
```

Let's see possible correlations between parameters and how 'good' they are. Low values of $\chi^2$ correspond to better solutions in the minimization process so they are preferred

```python
columns = np.array(RESULTS_opt).T

# Change here the parameters
parx, pary = 'argument2', 'mean_anomaly2'

x = columns[parameters[parx]+1].astype(np.float)
y = columns[parameters[pary]+1].astype(np.float)
c = np.log10(columns[0].astype(np.float))
f = columns[-1]

plt.figure(figsize=(12,12))

sc = plt.scatter(x, y, c=c, vmin=min(c) , vmax=max(c), cmap=plt.cm.rainbow, marker='o')

plt.colorbar(sc).set_label(r'$\log_{10}\chi^2$', fontsize=15)
plt.xlabel(f'{parx}'), plt.ylabel(f'{pary}');

#plt.savefig('DE_PO_NM_logchi2_2.png')
```

We can sort the results from the optimizers and select these solutions with lower $\chi^2$ values. 

```python
# We sort the results according to the chi square
sort_res = sorted(RESULTS_opt, key=lambda j: j[0])

# Let's take 'the best'
bestChi2 = sort_res[0][0]
best_planetary_params = sort_res[0][1:]

print(f"-Best chi**2 from the optimizers: {bestChi2}")
print("-Best planetary parameters: \n")
print("\n".join([ f"{p}: {best_planetary_params[i]}" for i, p in enumerate(PS.params_names_opt.split())]))
```

Let's plot the best n solutions from the optimizer and compare them with the observed TTVs. To do that, we import a useful function in the nauyaca package: plot_TTVs.

Pass the Planetary Object and a list of lists with the solutions you want to plot

```python
from nauyaca_local.plots import plot_TTVs

# Extract the best n solutions

n_sols = 1  # change here the number of best solutions from the optimizers!

best_n_solutions = [s[1:] for s in sort_res[:n_sols]]
print(best_n_solutions)
# Plot the observed TTVs and the 3 solutions
plot_TTVs(PS, best_n_solutions)

plt.savefig("TTVs_optimizer.png")
```

Ok, that does not looks like an ideal TTVs fit. Results from the optimizers provide an overall outlook of the solutions in the parameter space. To increase the chance of getting better solutions, just increase the number of solutions 'nsols' in Optimizers.run_optimizers. These solutions can be used to initialize walkers in the MCMC. We'll see how to use the information from the optimizers to start the MCMC in the next tutorial.
